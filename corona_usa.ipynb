### Anvender "drop()" funktionen s√• at vi kan fjerne nogen kolonner

```python
cleaned_data = data.drop(['UID','iso2','iso3','code3','FIPS','Admin2','Country_Region','Lat','Long_','Combined_Key'], axis = 1)
```

```python
cleaned_data.head()
```

# Gruppering: gruopby()

### Samler en stat sammen, vi grupper dem. groupby() funktionen, v√¶lger de stater med samme navne automatisk. 


```python
groupded_data = cleaned_data.groupby('Province_State').sum() # Summer hver stat sammen, ved at anvenden funktion 
```

```python
groupded_data.head()
```

```python
transposed_data = groupded_data.transpose() # Tranposer, s√• vi kan se alle staterne. 

transposed_data
```

## iloc[] : Tager indekset p√• et beskemt r√¶kke. Tallet i parentesen er ->
### 0 = F√∏rste r√¶kke, 1 = anden r√¶kke osv.
### -1 = sidste r√¶kke, -2  = anden sidste r√¶kke osv. 

```python
transposed_data.iloc[-1] # Tager den allersidste r√¶kke. I dette tilf√¶lde dato 10/4-2021. Antal smittet for hver stat. 
```

```python
plt.figure(figsize=(12, 5))
states = ['Texas', 'New York', 'Florida']# Bruger definerer de stater man gerne vil have
series = groupded_data.loc[states]
series = series.transpose() # Hiver transposed data ud og gemmer i variabel. 

series.index = pd.to_datetime(series.index) # Laver det til datoer objekter i stedet for en string. S√• bliver x-aksen p√¶nerer. 


plt.plot(series) # M√•den man laver linjeplot i matplotlib
#plt.yscale('log')

#states = series.columns # Gemmer kolonne navne i variabel
plt.legend(states)
plt.show()


```

```python
populations = {
    'Texas': 29000000,
    'Florida': 21500000,
    'New York': 19000000,
}
```

## Normaliser befolkningen 

```python
plt.figure(figsize=(12, 5))
states = ['Texas', 'New York', 'Florida']# Bruger definerer de stater man gerne vil have
series = groupded_data.loc[states]
series = series.transpose() # Hiver transposed data ud og gemmer i variabel. 

series.index = pd.to_datetime(series.index) # Laver det til datoer objekter i stedet for en string. S√• bliver x-aksen p√¶nerer. 

for state in states: 
    series[state] = series[state] / populations[state]



plt.plot(series) # M√•den man laver linjeplot i matplotlib
#plt.yscale('log')
plt.xlabel('Dato', fontsize = 16)
plt.ylabel('Antal tilf√¶lde', fontsize = 16)
#states = series.columns # Gemmer kolonne navne i variabel
plt.legend(states)
plt.savefig('states.png')
plt.show()

    
```
